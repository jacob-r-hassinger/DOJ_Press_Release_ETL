```python
import pandas as pd
import requests
from asset import ny_key
```

```python
url = "https://api.nytimes.com/svc/search/v2/articlesearch.json?"
```

```python
query =  "financial crime"

publication_range = "begin_date:2015-01-01&fq=end_date:2020-01-01"

query_url = url + "api-key=" + ny_key + "&q=" + query
```

```python
articles = requests.get(query_url).json()
```

```python
articles['response']['docs']
```

```python
#one articles table
articles_list = []
keywords_list = []
n=0
x=0
j=0
for article in articles["response"]["docs"]:
    articles_list.append(
        {"id":n+1,
         "abstract":articles['response']['docs'][n]['abstract'],
         "web_url":articles['response']['docs'][n]['web_url'],
         "snippet": articles['response']['docs'][n]['snippet'], 
         "lead_paragraph":articles['response']['docs'][n]['lead_paragraph']
        }
    )
    for keyword in articles['response']['docs'][n]['keywords']:
        keywords_list.append(
        {"id":j+1,
         "article_id":n+1,
         'name':articles['response']['docs'][n]['keywords'][x]['name'],
         'value':articles['response']['docs'][n]['keywords'][x]['value']}
         )
        x+=1
        j+=1
    n+=1
    x=0
```

```python
articles_df = pd.DataFrame(articles_list)
articles_df.head()
```

```python
keyword_df = pd.DataFrame(keywords_list)
keyword_df.head()
```
