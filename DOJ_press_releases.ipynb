```python
import pandas as pd
import json
```

```python
doj_press_releases = pd.read_json('doj_press_release.json', orient="columns", lines = True)
```

```python
doj_press_releases.head()
```

```python
doj_press_releases.describe()
```

```python
doj_topics = [topic for topic in doj_press_releases["topics"]]
```

```python
clean_topics = []
clean_topics_fk = []
clean_topics_pk = []
fk = 0
pk = 0
for topic in doj_topics:
    for n in topic:
        clean_topics.append(n)
        clean_topics_fk.append(fk+1)
        clean_topics_pk.append(pk+1)
        pk+=1
    fk+=1
```

```python
doj_topics_df = pd.DataFrame({"ID":clean_topics_pk,
                              "Press_Release_ID":clean_topics_fk,
                              "Topic":clean_topics})
doj_topics_df.head()
```

```python
doj_topics_df.describe()
```

```python
doj_topics = [component for component in doj_press_releases["components"]]
clean_components = []
clean_components_fk = []
clean_components_pk = []
fk = 0
pk = 0
for topic in doj_topics:
    for n in topic:
        clean_components.append(n)
        clean_components_fk.append(fk+1)
        clean_components_pk.append(pk+1)
        pk+=1
    fk+=1
doj_components_df = pd.DataFrame({"ID":clean_components_pk,
                              "Press_Release_ID":clean_components_fk,
                              "Component":clean_components})
doj_components_df.head()
```

```python
id = 0
ids = []
for title in doj_press_releases["title"]:
    ids.append(id+1)
    id+=1
doj_id_df = pd.DataFrame({"ID":ids})
```

```python
doj_press = doj_press_releases[["title", "contents", "date"]]
doj_id_df.head()
```

```python
doj_press["id"] = doj_id_df["ID"]
```

```python
doj_press = doj_press[["id", "title", "contents", "date"]]
doj_press.head()
```
